```
df=pd.read_csv(r'/content/IPL2025Batters (1).csv')
df.head()
```

```
df.Matches=df.Matches.fillna('Zero')
df
```

```
df.Matches=df.Matches.apply(w2n.word_to_num)
df
```

```
import math
median_test_score=math.floor(df['Runs'].mean())
median_test_score
```

```
dff=df['Runs'].mean()
dff
```

```
df['Matches']=df['Matches'].fillna(dff)
df
```

```
df.columns
```

```
result=['Team',	'Runs',	'Matches']
x=df[result]
y=df['Runs']
```

```
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```
x_train.shape,x_test.shape
```

```
y_train.shape,y_test.shape
```

```
reg=LinearRegression()
```

```
model=reg.fit(x_train,y_train)
```

```
model.predict([[7,10,9]])
```

```
model.score(x_train,y_train)
```

```
model.score(x_test,y_test)
```
